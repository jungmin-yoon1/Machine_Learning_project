```python
import numpy as np # 벡터 및 행렬 연산을 위함
import pandas as pd # 데이터 분석에 활용됨
import matplotlib.pyplot as plt # 그래프를 그리기 위함
import random
```

```python
np.random.seed(777)

mu1 = [7.0, -1.0]
mu2 = [3.0, -1.5]
mu3 = [5.5, 1.0]
sigma = np.array([[0.7, 0], [0, 0.7]])

c1 = np.random.multivariate_normal(mu1, sigma, 100)  #cluster 1
c2 = np.random.multivariate_normal(mu2, sigma, 100)  #cluster 2
c3 = np.random.multivariate_normal(mu3, sigma, 100)  #cluster 3

c_1 = np.concatenate([c1, c2]) 
Dataset = np.concatenate([c_1, c3]) #전체 dataset

#Plot the generated dataset in 2D scatter plot
plt.scatter(c1[:, 0], c1[:, 1])
plt.scatter(c2[:, 0], c2[:, 1])
plt.scatter(c3[:, 0], c3[:, 1])
plt.show()
```

# K-means algorithm

K-means: K=2

```python
m = Dataset.shape[0] 
n = Dataset.shape[1] 
n_iter=100

# Empty centroid array 생성
centroids=np.array([]).reshape(n,0)  

K=2 #K cluster로 나눔

distortion = []

# K random centroids 생성
for k in range(K):
    centroids=np.c_[centroids,Dataset[random.randint(0,m-1)]]
    
for i in range(n_iter):
      euclid=np.array([]).reshape(m,0)
      for k in range(K):
          dist=np.sum((Dataset-centroids[:,k])**2,axis=1)
          euclid=np.c_[euclid,dist]
      minimum=np.argmin(euclid,axis=1)+1

      cent={}
      for k in range(K):
           cent[k+1]=np.array([]).reshape(2,0)
      for k in range(m):
           cent[minimum[k]]=np.c_[cent[minimum[k]],Dataset[k]]
      for k in range(K):
           cent[k+1]=cent[k+1].T
      for k in range(K):
           centroids[:,k]=np.mean(cent[k+1],axis=0)
      final=cent  
    
      #같은 cluster에 속하는 각각의 점들로부터 그 cluster의 평균(centroid)과의 거리의 합의 제곱의 평균을 구해 Distortion 저장
      D = 0
      for k in range(K):
        dist_cluster = np.mean(np.sum((final[k+1]-centroids[:,k])**2,axis=1))
        D = D + dist_cluster
      distortion.append(D/K)

#Plot the final centroids over the dataset plot
plt.figure(figsize = (13, 4))
plt.subplot(1,2,1)
for k in range(K):
    plt.scatter(final[k+1][:,0],final[k+1][:,1])
   
plt.scatter(centroids[0,:],centroids[1,:],s=200, c='k', marker='x', linewidth='3')
plt.title('K=2')

#Provide traing Curve every iterations
plt.subplot(1,2,2)
plt.title('Training Curve')
plt.plot(distortion)
plt.xlabel('Training iteration')
plt.ylabel('Distortion')
plt.legend()
plt.show()
```

#GMM algorithm

GMM class 생성

```python
import numpy as np
from scipy.stats import multivariate_normal

class GMM:
    def __init__(self, k, max_iter=5):
        self.k = k
        self.max_iter = int(max_iter)

    def initialize(self, X):
        self.shape = X.shape
        self.n, self.m = self.shape

        self.phi = np.full(shape=self.k, fill_value=1/self.k)
        self.weights = np.full( shape=self.shape, fill_value=1/self.k)
        
        random_row = np.random.randint(low=0, high=self.n, size=self.k)
        self.mu = [  X[row_index,:] for row_index in random_row ]
        self.sigma = [ np.cov(X.T) for _ in range(self.k) ]
        

    def e_step(self, X):
        # E-Step: update weights and phi holding mu and sigma constant
        self.weights = self.predict_proba(X)
        self.phi = self.weights.mean(axis=0)
    
    def m_step(self, X):
        # M-Step: update mu and sigma holding phi and weights constant
        for i in range(self.k):
            weight = self.weights[:, [i]]
            total_weight = weight.sum()
            self.mu[i] = (X * weight).sum(axis=0) / total_weight
            self.sigma[i] = np.cov(X.T, 
                aweights=(weight/total_weight).flatten(), 
                bias=True)

    def fit(self, X):
        self.initialize(X)
        
        for iteration in range(self.max_iter):
            self.e_step(X)
            self.m_step(X)
        
        mu_final = self.mu
        return mu_final

            
    def predict_proba(self, X):
        likelihood = np.zeros( (self.n, self.k) )
        for i in range(self.k):
            distribution = multivariate_normal(
                mean=self.mu[i], 
                cov=self.sigma[i])
            likelihood[:,i] = distribution.pdf(X)
        
        numerator = likelihood * self.phi
        denominator = numerator.sum(axis=1)[:, np.newaxis]
        weights = numerator / denominator
        return weights
    
    def predict(self, X):
        weights = self.predict_proba(X)
        return np.argmax(weights, axis=1)
```

GMM: K=2  

```python
gmm = GMM(k=2, max_iter=100) # k = 2 (Cluster 수)
mean = gmm.fit(Dataset)  #각 cluster의 평균 값
labels = gmm.predict(Dataset)

mean_array = np.array(mean)  
mean_array.reshape(-1, 2)  #mean을 배열형태로 변경

#각 label로 분류된 2개의 cluster의 데이터 저장
d0 = [] 
d1 = [] 

for i in range(len(labels)):
  if labels[i]==0:
    d0.append(list(Dataset[i]))
  elif labels[i] == 1:
    d1.append(list(Dataset[i]))

#배열형태로 변경
d0_array = np.array(d0) 
d0_array.reshape(-1, 2)

d1_array = np.array(d1)
d1_array.reshape(-1, 2)
 
# Plot two clusters in same plot
plt.scatter(d0_array[:, 0], d0_array[:, 1])
plt.scatter(d1_array[:, 0], d1_array[:, 1])

# Plot the obtained mean
plt.scatter(mean_array[:, 0],mean_array[:, 1],s=200, marker='x', c='k', linewidth='3')
plt.show()
```
